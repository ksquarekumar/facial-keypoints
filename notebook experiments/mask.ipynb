```python
import os
import cv2
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
os.getcwd()
```

```python
image = cv2.cvtColor(cv2.imread('stop_pm.jpg'), cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
lower_red = np.array([112,0,0]) 
upper_red = np.array([255,96,96])
```

```python
masked_image = np.copy(image)
mask = cv2.inRange(masked_image, lower_red, upper_red)
mask = cv2.dilate(mask, np.ones((2,2)), 2)


plt.imshow(mask, cmap='gray')
```

```python
maskedimg = np.copy(image)
maskedimg[mask!=0] = [0,0,0]
```

```python
plt.imshow(maskedimg)
```
