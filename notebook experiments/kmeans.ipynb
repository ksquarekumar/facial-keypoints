```python
import os
import cv2
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
image = cv2.imread('balloon.jpg')
```

```python
plt.imshow(image[:,:,::-1])
```

```python
imagecopy = np.copy(image)
pixel_vals = np.float32(imagecopy.reshape((-1,3)))
# how many channels to split into
k = 16
# tells when to stop
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
retval, labels, centers = cv2.kmeans(pixel_vals.reshape((-1,3)),
                                     k,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
```

```python
centers = np.uint8(centers)
seg_data = centers[labels.flatten()]
```

```python
seg_image = seg_data.reshape((image.shape))
```

```python
labels_reshape = labels.reshape(image.shape[0], image.shape[1])
```
